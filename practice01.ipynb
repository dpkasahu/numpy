2.Given arr = np.array([5, 10, 15, 20, 25, 30, 35]): Extract elements from index 2 to 5 and reverse

6. Create a 1D array of numbers 1 to 15 and extract every second element and every third element
starting from index 1

```python
arr=np.arange(1,16)
print("Every 2nd element:",arr[::2])
print("Every 3rd element from index 1:", arr[1::3])
```
