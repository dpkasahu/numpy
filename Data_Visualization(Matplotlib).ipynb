```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=x**2

#plt.plot(x,y)
plt.plot(x,y,marker='P',color='red',markerfacecolor='white',markersize=15,markeredgecolor='black',linestyle='--')

#to use plot the length of x and y must be same.


```

```python
x=np.array([1,2,3,4,5])
y=x**2
z=x**3

#plt.plot(x,y)
plt.plot(x,y,marker='h',color='red',markerfacecolor='white',markersize=15,markeredgecolor='black',linestyle='--',label='square')
plt.title('number with squares',c='blue')
plt.xlabel('x-axis',c='black')
plt.ylabel('y-axis',c='orange')
plt.plot(x,z,color='black',label='cube')
plt.legend()
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=x**2
z=x**3
a=x**4

#plt.plot(x,y)
plt.plot(x,y,marker='h',color='red',markerfacecolor='white',markersize=15,markeredgecolor='black',linestyle='--',label='square')
plt.title('number with squares',c='blue')
plt.xlabel('x-axis',c='black')
plt.ylabel('y-axis',c='orange')
plt.plot(x,z,color='black',label='cube')
plt.plot(x,a,color='green',label='fourth')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])

plt.scatter(x,y,marker ='*',s=50)
plt.gca().set_facecolor('yellow')
  plt.annotate(f'({x[4]},{y[4]})',(x[4],y[4]))
  plt.annotate(f'({x[3]},{y[3]})',(x[3],y[3]))
  plt.annotate(f'({x[2]},{y[2]})',(x[2],y[2]))
  plt.annotate(f'({x[1]},{y[1]})',(x[1],y[1]))
  plt.annotate(f'({x[0]},{y[0]})',(x[0],y[0]))

plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])

plt.scatter(x,y,marker ='*',s=50)
plt.gca().set_facecolor('yellow')
plt.xlabel('x-axis',c='red')
plt.ylabel('y-axis',c='orange')
plt.title('scatter plot')
for i in range(0,len(x)):
  plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]))
plt.show()
```

1. Bar Graph:
---------------

A bar graph uses separate rectangular bars to show and compare quantities of different categories.

Example: Comparing the number of students in different classes.

2. Histogram:
------------

A histogram looks like a bar graph but is used for continuous data.

The bars touch each other because the data represents ranges (intervals).

Example: Showing the distribution of marks (010, 1020, 2030...).

3. Pie Chart:
------------

A pie chart is a circular chart divided into slices. Each slice shows a proportion of the whole.

Example: Showing percentage of expenses like food, travel, rent, etc.

4. Scatter Plot:
----------------

A scatter plot shows dots on a graph to represent the relationship between two variables.

Helps in finding patterns, trends, or correlations.

Example: Height vs weight of students.

ydifferences between scatter,bar,histogram,pie chart.

1. Bar Graph

Used for categorical data (names, types, groups).

Bars are separate (no touching).

Shows comparison between different groups.

Example: Number of boys and girls in a class.

2. Histogram

Used for continuous numerical data (ranges/intervals).

Bars are touching because data is in continuous intervals.

Shows distribution or spread of data.

Example: Distribution of ages (010, 1020).

3. Pie Chart

Used to show parts of a whole (percentages).

Data is shown as slices of a circle.

Good for showing proportions, not exact values.

Example: Percentage of expenses in a budget.

4. Scatter Plot

Used to show the relationship between two variables.

Data points are dots, not bars or slices.

Helps identify trends, correlation (positive/negative).

Example: Study hours vs marks.
