```python
import pandas as pd

s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series([1, 2, 3, 4])

print("Series 1:\n", s1)
print("Series 2:\n", s2)
```

```python
print("Addition:\n", s1 + s2)
print("Subtraction:\n", s1 - s2)
print("Multiplication:\n", s1 * s2)
print("Division:\n", s1 / s2)
print("Floor Division:\n", s1 // s2)
print("Modulo:\n", s1 % s2)
print("Power:\n", s1 ** s2)
```

```python
print("Greater Than:\n", s1 > s2)
print("Less Than:\n", s1 < s2)
print("Equal:\n", s1 == s2)
print("Not Equal:\n", s1 != s2)
```

```python
cond1 = s1 > 15     # True where s1 > 15
cond2 = s2 < 4      # True where s2 < 4

print("Logical AND:\n", cond1 & cond2)
print("Logical OR:\n", cond1 | cond2)
print("Logical NOT on cond1:\n", ~cond1)
```
