```python
py_list=[1,2,3]
print(py_list * 2)
```

```python
import time
start = time.time()
py_list=[i*2  for i in range(10000000)]
print("list:", time.time()-start)
start=time.time()
np_array=np.array([i*2 for i in range(10000000)])
print("numpy array:",time.time()-start)
```

```python
import numpy as np
import matplotlib.pyplot as plt

sales_data = np.array([
    [1,150000,180000,220000,250000],
    [2,120000,140000,160000,190000],
    [3,120000,280000,290000,190000],
    [4,130000,360000,170000,150000],
    [5,160000,185000,205000,230000]
])
print("===Zomato sales analysis===")
sales_data.shape
sales_data[:3]
```

Total sales per year

```python
# print(np.sum(sales_data,axis = 0))
yearly_total = np.sum(sales_data[:,1:],axis = 0)
print(yearly_total)
```

Min sales per resturant

```python
min_sales = np.min(sales_data[:,1:],axis = 1)
print(min_sales)
```

Maximun sales per year

```python
max_sales = np.max(sales_data[:,1:],axis = 0)
print(max_sales )
```

Average sales per resturant

```python
avg_sales = np.mean(sales_data[:,1:],axis = 0)
print(avg_sales)
```

cumulative sales

```python
cumsum = np.cumsum(sales_data[:,1:],axis = 0)
print(cumsum)
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis = 0))
plt.title("Average cumulative sales across all resturant ")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```
