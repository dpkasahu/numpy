```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/whoknowayush/DataAnalysis-Dataset/refs/heads/main/Datasets-Practice/employee_data.csv')
df
```

```python
df.columns = df.columns.str.upper()
df
```

```python
df["JOIN_DATE"] = pd.to_datetime(df["JOIN_DATE"])
df["EMP_STATUS"] = df["EMP_STATUS"].str.lower()
```

```python
df["ACTIVE"] = df["EMP_STATUS"] == "permanent"
```

```python
df[
    (df["ACTIVE"] == True) &
    (df["EXP_YEARS"] >= 5) &
    (df["SALARY"] > 50000)
]
```

```python
df
```

```python
df["YEARS_WITH_COMPANY"] = 2025 - df["JOIN_DATE"].dt.year
```

```python
df["SENIOR"] = df["EXP_YEARS"] >= 10
```

```python
df["FIRST_NAME"] = df["FULL_NAME"].str.split().str[0]
df["LAST_NAME"] = df["FULL_NAME"].str.split().str[-1]
```

```python
df["EXP_AGE_RATIO"] = df["EXP_YEARS"] / df["AGE"]
```

```python
df["NAME_LENGTH"] = df["FULL_NAME"].str.len()
```

```python
df["DEPT_SENIORITY"] = df["DEPARTMENT"] + "_" + df["SENIOR"].astype(str)
```

```python
df["SALARY_RELATIVE"] = df["SALARY"] / df.groupby("DEPARTMENT")["SALARY"].transform("mean")
```

```python
df =df.set_index("EMP_ID")
```

```python
df = df.reset_index()
df
```

```python
df.to_csv('emp_cleaned')
```
