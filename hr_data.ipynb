```python
import pandas as pd
```

```python
df1=pd.read_csv('https://raw.githubusercontent.com/whoknowayush/DataAnalysis-Dataset/refs/heads/main/Datasets-Practice/hr_data.csv')
df1.info()
```

```python
df1.groupby('salary').get_group('high').groupby('department').get_group('sales').groupby('number_project').get_group(2)
```

```python
df1.groupby('department').get_group('support').groupby('number_project').get_group(2)['time_spend_company'].sum()
```

```python
df1.groupby('salary').get_group('low').groupby('time_spend_company').get_group(3)['average_montly_hours'].sum()
```

```python
df1.pivot_table(values=['average_montly_hours','number_project','time_spend_company'],index='department',aggfunc=[max,min])
```
