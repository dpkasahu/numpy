1)distplot:  sns.distplot('col_name',bins=,kde=)
----------
how many times the value occured.

kde=>kernal density estimator.

if kde=True
y-axis represents density.

if kde=False
y-axis represents count.



```python
sns.distplot(sb1['sepal_length' ])
```

```python
sns.distplot(sb1['sepal_length' ],bins=30)
```

```python
sns.distplot(sb1['sepal_length' ],bins=30,kde=False) #false ->density
```

```python
sns.distplot(sb1['sepal_length' ],bins=30,kde=True,color='green') #true ->curve

```

```python
sns.distplot(sb1['petal_length' ],bins=30,kde=True,color='orange')
```

```python
sns.distplot(sb1['sepal_width' ],bins=30,kde=True,color='brown')
```

```python
sns.distplot(sb1['petal_width' ],bins=30,kde=True,color='black')
```

```python
sns.jointplot(x='sepal_width',y='sepal_length',data=sb1,hue='species')
```

2)Categorical plot:

To deal with the categorical columns we use categorical plot.

1)count plot.

4)box plot.
5)violin plot.





1)count plot:
------------

It will count the data points of the categorical column.

syntax:-

sns.countplot(x='car_col',data=sb)

```python
import pandas as pd
import seaborn as sns
sb=pd.read_csv('https://raw.githubusercontent.com/whoknowayush/DataAnalysis-Dataset/refs/heads/main/Datasets-Practice/tips.csv')
```

```python
sns.countplot(x='day',data=sb)
```

```python
a=sns.countplot(x='day',data=sb)
for i in a.containers: #sun sat thur fri (bars)
  a.bar_label(i)
```

```python
b=sns.countplot(x='day',data=sb,hue='sex')
for i in b.containers:
  b.bar_label(i)
```

```python
c=sns.countplot(x='sex',data=sb,color='lavender')
for i in c.containers:
  c.bar_label(i)
```

```python
e=sns.countplot(x='time',data=sb,color='brown')
for i in e.containers:
  e.bar_label(i)
```

```python
sb.info()
```

```python
f=sns.countplot(x='smoker',data=sb,color='green')
for i in f.containers:
  f.bar_label(i)
```

Matrix plot
----
it is used to find the relation between multiple columns in matrix format.

for this we should have an idea of correlation efficient.

data.corr() can be applied on numerical columns. it will give the range between -1 to +1 , whereas -1 is for negatively charged and +1 for positively charged.

if we have categorical column perform slicing.

```python
import seaborn as sns
import pandas as pd
sb=sns.load_dataset('iris')
sb
```

```python
df=sb.iloc[:,0:4]
cor=df.corr()
sns.heatmap(cor)
# cor = sb.corr()
sns.heatmap(cor,annot=True)
```

```python
import seaborn as sns
import pandas as pd
pv = pd.pivot_table(values=['sepal_length','sepal_width','petal_length','petal_width'],index='species',data=sb,aggfunc=[max])
pv
```

```python
sns.heatmap(pv,annot=True)
```
