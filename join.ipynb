```python
import pandas as pd
```

```python
employee=pd.DataFrame({'emp_name':['Deepika','yamini','deeya'],
                      'emp_id': [101,102,103]})
salary = pd.DataFrame({'emp_id':[101,102,104,105],
                       'emp_sal':[60000,45000,70000,50000]})
pd.concat([employee,salary])
```

```python
#inner #outer #left #right
pd.merge(employee,salary,how='inner',on='emp_id')
#only will give common rows
```

```python
pd.merge(employee,salary,how='outer',on='emp_id')
#return both matching and non-matching rows
```

```python
pd.merge(employee,salary,how='left',on='emp_id')
#return all rows of left side both matching and non-matching and  common of right side
```

```python
pd.merge(employee,salary,how='right',on='emp_id')
#return matching and non-matching of right side and matching of left side rows
```
