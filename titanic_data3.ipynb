```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/whoknowayush/DataAnalysis-Dataset/refs/heads/main/Datasets-Practice/titanic.csv')
df.head()
```

```python
# df['title'] = df['name'].str.split(',').str[1].str.split('.').str[0].str.strip()
df['title'] = [name.split(',')[1].split('.')[0].strip() for name in df['name']]
df['fname'] = df['name'].str.extract(r'\.\s*([A-Za-z]+)')[0]
df
```
