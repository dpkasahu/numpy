COPY OPERATION

```python
#normal copy = if u change the destination variable it will change the source variable
#dest_variable = src_variable (syntax)
#in this copy operation when src_variable memory address will get copied to dest_variable
a=np.array([1,2,3,4,5])
b=a
a[2]=30
b
```

```python
#shallow copy = if you change the destination variable it will not change the source variable
#it is an copy operation when the value space will get copied to dest_variable
#dest_var = src_var.copy()
dest_var = src_var.copy()
b = np.array([1,2,3,4,5])
c=b.copy()
c[2]=20
b
```

DIFFERENCE BETWEEN VIEW() AND NORMAL COPY.

```python
#IN NORMAL COPY SRC_VAR AND DEST_VAR WILL HAVE SAME ADDRESS.
#IN view() src_var and dest_var will have different address.
```

```python
#view()
d=np.array([1,2,3,4,5])
e=d.view()
e[2]=30
d
```
