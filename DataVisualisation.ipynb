```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=x**2
plt.plot(x,y,marker='p',color='brown',markerfacecolor='blue',markersize=10,markeredgecolor='black')
#to use plot the length of x and y must be same
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = x**2
z=x**3
a=x**4
# plt.scatter(x, y)
# plt.bar(x, y)
# plt.plot(x, y, linestyle='--', marker='o')
# plt.show()
plt.fill_between(x, y, color='skyblue', alpha=0.3, label='Square')
plt.plot(x, z, color='orange', alpha=0.3 , label='Cube')
plt.plot(x, a, color='black', alpha=0.3 , label='PowerFour')
plt.legend()
# plt.title()
plt.plot(x, y)
plt.plot(x, z,label='cube')
plt.xlabel('Numbers',c='brown')
plt.ylabel('Squares',c='brown')
```
