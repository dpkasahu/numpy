Distribution plot :-

if datatype == object ,
it will throw new error

1)displot

```python
import seaborn as sns
```

```python
sb1=sns.load_dataset('iris')
sb1
```

```python
sns.distplot(sb1['sepal_length'],bins=20)
```

```python
sns.distplot(sb1['sepal_length'],bins=20,kde=False,color='green') #kde = kernel density estimator , False = show density y-axis show count , true = default value y-axis show density
```

```python
sns.distplot(sb1['sepal_width'],bins=30,kde=False,color='red')
```

```python
sns.distplot(sb1['petal_width'],bins=30,kde=True,color='green')
```

```python
sns.jointplot(x='sepal_width',y='sepal_length',data=sb1,kind="hex") #kind = "kde","hex"
```

```python
sns.jointplot(x='sepal_width',y='sepal_length',data=sb1,hue='species')
```
